```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dataset/datasets_top_4.csv', index_col = 'dataset')
df
```

```python
df1 = df.iloc[0:13]
df2 = df.iloc[13:]
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)

fig.set_size_inches(7, 4)

df1.plot.barh(
    stacked = True,
    ax = ax1,
    # Ancho de las columnas 
    width = 0.3,
    color = {
        'cat1' : '#000086',
        'cat2' : '#5396f4',
        'cat3' : '#a1dbe8',
        'otros' : '#DADADA'
    }
)

df2.plot.barh(
    stacked = True,
    ax = ax2,
    width = 0.3,
    color = {
        'cat1' : '#000086',
        'cat2' : '#5396f4',
        'cat3' : '#a1dbe8',
        'otros' : '#DADADA'
    }
)

ax2.set_ylabel('')
ax1.set_ylabel('Dataset')

ax1.set_xlabel('Porcentaje')
ax2.set_xlabel('Porcentaje')

ax1.legend().remove()

ax2.legend(
    ['Categoria 1', 'Categoria 2', 'Categoria 3', 'Categoria 4'],
    ncol = 4,
    loc = 'upper left',
    bbox_to_anchor = (-1.5, -.12)
)

# mover las etiquetas al lado derecho
ax2.yaxis.tick_right()

#etiqueta debajo del grafico
ax1.text(
    30, -4,
    'Algunos datasets han sido ignorados del analisis',
    bbox = {
        'facecolor' : 'orange',
        'alpha' : 0.5
    }
)

fig.suptitle('Como se distribuyen las categorias dentro de los datasets?')
```
