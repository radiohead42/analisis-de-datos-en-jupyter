```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('comparasion_algoritmos.csv')
df
```

```python
#plt.style.use('ggplot')

ax = df.plot(
    kind = 'bar',
    x = 'dataset',
    y = ['algoritmo1', 'algoritmo2'],
    figsize = (12, 5)
)

ax.grid(visible = True)

ax.set_xlabel('Dataset')

ax.set_ylabel('Rendimiento')

ax.set_title('Evaluacion de Algoritmo1 VS Algoritmo2')

ax.legend(
    title = '',
    labels = ['Algoritmo1', 'Algoritmo2']
)

ax.set_axisbelow(True)
```
